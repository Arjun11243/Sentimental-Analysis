```python
import nltk
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
import pandas as p
```

```python
nltk.download('movie_reviews')
```

```python
from nltk.corpus import movie_reviews,stopwords
reviews=[]
for fileid in movie_reviews.fileids():
    review=' '.join(movie_reviews.words(fileid))
    reviews.append((review,fileid[:3]))
dataset=p.DataFrame(reviews,columns=['Review','Label'])
dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(dataset['Review'],dataset['Label'],test_size=0.3,random_state=42)
X_train,X_test,Y_train,Y_test
```

```python
vectorizer=CountVectorizer()
X_train_counts=vectorizer.fit_transform(X_train)
X_test_counts=vectorizer.transform(X_test)
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train_counts,Y_train)
```

```python
logistic_prediction=logistic.predict(X_test_counts)
```

```python
for i,j in zip(X_test,logistic_prediction):
    print(i[:10]," | ",j)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y_test,logistic_prediction)
accuracy
```

```python
text=["This Movie is absolutely great"]
c1=vectorizer.transform(text)
logistic.predict(c1)
```

# **Bayes**

```python
from sklearn.naive_bayes import MultinomialNB
bayes=MultinomialNB()
bayes.fit(X_train_counts,Y_train)
```

```python
bayes_prediction=bayes.predict(X_test_counts)
```

```python
for i,j in zip(X_test,bayes_prediction):
    print(i[:10]+" | "+j)
```

```python
accuracy=accuracy_score(Y_test,bayes_prediction)
accuracy
```

```python
text=["This moovie is absolutely best"]
c2=vectorizer.transform(text)
bayes.predict(c2)
```
